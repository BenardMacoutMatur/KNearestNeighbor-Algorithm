```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import seaborn as sns
import openai as oi
from sklearn.preprocessing import StandardScaler #preprocessor
from sklearn.neighbors import KNeighborsClassifier#ktool
from sklearn.metrics import confusion_matrix #testing model
from sklearn.metrics import f1_score #testing model- how good is the model
from sklearn.metrics import accuracy_score #testing also
```

```python
df=pd.read_csv("Downloads\\diabetes.csv")
df
```

```python
zero_not_accepted=['Glucose','BloodPressure','SkinThickness','BMI','Insulin']
#list=[]
#tuple=()
#set={}
#dictionary={}

for column in zero_not_accepted:
    df[column]=df[column].replace(0,np.NaN)
    mean=int(df[column].mean(skipna=True))
    df[column]=df[column].replace(np.NaN, mean)
```

```python
#Split the data
#df.loc[#rows,:columns]
X=df.iloc[:,0:8]
y=df.iloc[:,8]
X_test,X_train,y_test,y_train=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
#FEATURE SCALING.
sc_X=StandardScaler()
X_train=sc_X.fit_transform(X_train)
X_test=sc_X.transform(X_test)
```

```python
 classifier=KNeighborsClassifier(n_neighbors=11,p=2,metric='euclidean')
```

```python
#fitmodel
classifier.fit(X_train,y_train)
```

```python
#predict the test set results.
y_pred=classifier.predict(X_test)
y_pred
```

```python
#evaluate the model
cm=confusion_matrix(y_test,y_pred)
cm
```

```python
f1_score(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```
